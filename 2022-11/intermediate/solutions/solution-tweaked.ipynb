```python
import re
from collections import Counter
```

```python
list_toplevel = Counter()
list_fulldomain = Counter()
list_postcode = Counter()
list_gender = Counter()

# top-level domain
myRegex1 = re.compile(r'\.[a-z]{1,3}\s')
# users in each full domain
myRegex2 = re.compile(r'[a-z0-9.]*\.[a-z.]{1,3}')
# postcode
myRegex3 = re.compile(r'([0-9]+)([-0-9]+){0,1}$')
# gender
myRegex4 = re.compile(r"Male|Female", re.IGNORECASE)

for line in open("fake_client_data.csv", 'r'):
    for x in myRegex1.findall(line):
        list_toplevel[x]+=1
    for x in myRegex2.findall(line):
        list_fulldomain[x]+=1
    for x in myRegex3.findall(line):
        list_postcode[x[0]]+=1
    for x in myRegex4.findall(line):
        list_gender[x.capitalize()]+=1 #remove '.capitalize()' to see what variation in values we are dealing with
```

```python
print ("The list for top level domains sorted by frequency")
list_toplevel.most_common()
#list_toplevel.most_common(10) #to get the top 10
```

```python
print ("The list of full domain names sorted by frequency")
list_fulldomain.most_common()
```
